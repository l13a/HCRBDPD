```python
import scglue
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout

# Load graph
grn = scglue.genomics.read_ctx_grn("../../data/GRN_Mic/HC_pruned_grn.csv")

# Set canvas
plt.figure(figsize=(60, 60))

# Compute layout
pos = graphviz_layout(grn)

# Draw nodes (with proper size)
nx.draw_networkx_nodes(
    grn,
    pos=pos,
    node_size=8000,
    node_color='skyblue',
    edgecolors='black'
)

# Draw labels (with font size)
nx.draw_networkx_labels(
    grn,
    pos=pos,
    font_size=25
)

# Optional: draw edges
nx.draw_networkx_edges(grn, pos=pos, width = 10)

# Save to file
plt.axis("off")
plt.savefig("../../data/GRN_Mic/HC_pruned_grn.pdf", format="pdf", bbox_inches="tight")
plt.close()

# Get node labels
labels_HC = list(grn.nodes())
print(labels_HC)
```
