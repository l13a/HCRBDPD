```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
```

```python
# Load graph
grn = scglue.genomics.read_ctx_grn("../../data/GRN_Inh_subset/HC_pruned_grn.csv")

# Set canvas
plt.figure(figsize=(60, 60))

# Compute layout
pos = graphviz_layout(grn)

# Draw nodes (with proper size)
nx.draw_networkx_nodes(
    grn,
    pos=pos,
    node_size=8000,
    node_color='skyblue',
    edgecolors='black'
)

# Draw labels (with font size)
nx.draw_networkx_labels(
    grn,
    pos=pos,
    font_size=25
)

# Optional: draw edges
nx.draw_networkx_edges(grn, pos=pos, width = 10)

# Save to file
plt.axis("off")
plt.savefig("../../data/GRN_Inh_subset/HC_pruned_grn.pdf", format="pdf", bbox_inches="tight")
plt.close()

# Get node labels
labels_HC = list(grn.nodes())
print(labels_HC)
```

```python
import pandas as pd

# Path to the Excel workbook
excel_path = "../../DEG_GO_Analysis/DEGs_RBD_vs_HC/Significant_DEGs_RBD_vs_HC_Inh_subset.xlsx"

# Collect results in one place
hits = []

# Load the workbook once
xls = pd.ExcelFile(excel_path)

# Iterate over all subtype-specific sheets
for sheet in xls.sheet_names:
    # Read the current sheet
    df = pd.read_excel(xls, sheet_name=sheet)
    
    # Assume first column holds the DEG gene names
    gene_col = df.columns[0]
    
    # Keep only rows whose gene is in your label list
    overlap = df[df[gene_col].isin(labels_RBD)]
    
    # Record matches with subtype and effect size
    for _, row in overlap.iterrows():
        hits.append(
            {
                "gene"      : row[gene_col],
                "subtype"   : sheet,
                "avg_log2FC": row["avg_log2FC"],
            }
        )

# Turn into a tidy table and print
overlap_df = pd.DataFrame(hits)
print(overlap_df.to_string(index=False))
```
